```python
import pandas as pd
columns = ['Date','Open','High','Low','Close','Volume','Yield','PercentageVolume',
           'SMA6','EMA6','WMA6','HMA6','SMA20','EMA20','WMA20','HMA20','SMA50','EMA50','WMA50','HMA50',
           'SMA100','EMA100','WMA100','HMA100','MACD','CCI','Stochastic Oscillator','RSI','ROC','PPO',
          'KST','BOLU','BOLD','BOLM']
df = pd.read_csv('DataReady/EURUSD/EURUSD_D1.csv', header =0, names = columns)
df = df.drop('Date', axis = 1)
df
```

```python
df2 = df.diff() 
df2.dropna(how='any')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (30,10))
plt.plot(range(len(df['Close'])), df['Close'])
plt.grid()
plt.show()

plt.figure(figsize = (30,10))
plt.plot(range(len(df2['Close'])), df2['Close'], label = 'Close Derivative')
plt.grid()
plt.show()
```
